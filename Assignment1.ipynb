14. How is memory managed in Python? 
"In Python, memory management is automatic and is handled by the Python interpreter's memory manager. The memory manager is responsible for allocating and deallocating memory for objects, as well as handling memory allocation for low-level data types like integers and strings.

Python uses a reference counting system to track the number of references to an object. When the reference count for an object reaches zero, the memory occupied by that object is automatically deallocated.

Additionally, Python also has a garbage collector that periodically frees up memory that is no longer being used. The garbage collector keeps track of objects that are still in use, and reclaims memory occupied by objects that are no longer needed.

In summary, Python's memory management is efficient, flexible, and takes care of the low-level details of memory management, freeing up the developer to focus on writing code.

16. Is indentation required in python? 
"Yes, indentation is required in Python. Indentation is used to define blocks of code, such as loops and functions. In Python, indentation is used instead of curly braces {} or keywords like begin and end, which are used in other programming languages.

For example, a loop in Python is defined like this:

      for i in range(10):
      print(i)
      
In this example, the code inside the loop is indented four spaces, indicating that it is part of the loop. In Python, a consistent indentation style is important, as indentation determines the structure of the code. Typically, a standard indentation size of four spaces is used in Python, but it can be any number of spaces as long as it is consistent within a single code block.
